```python
# Standard library imports.
from urllib.request import urlopen
from json import load

# Third-party imports.
from pandas import DataFrame

# The URL, including the predetermined API key.
url = ("https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json"
       "?api-key=cc517d66ab6c4e4dab0605bf4d50bd32")
       
# The list of fields to include per results, omitting Several fields that are
# either lists or frequently null.
fields = ["title", "description", "publisher", "author", "price"]

# Read the URL and place its contents into a data frame with the given fields.
DataFrame(load(urlopen(url))["results"], columns=fields)
```
